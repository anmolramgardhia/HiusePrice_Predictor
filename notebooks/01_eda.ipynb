# EDA — Ames Housing Dataset

This notebook covers:
1. Dataset overview and shape
2. Target variable (SalePrice) distribution
3. Correlation heatmap — top 15 features
4. Missing value analysis
5. Key feature distributions (GrLivArea, OverallQual, Neighborhood)
6. Outlier identification

```python
import sys
sys.path.insert(0, '../src')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from data import load_train, remove_outliers

%matplotlib inline
sns.set_theme(style='whitegrid', palette='muted')
plt.rcParams['figure.figsize'] = (12, 5)
```

```python
df = load_train()
print(df.shape)
df.head()
```

```python
# Target distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 4))

axes[0].hist(df['SalePrice'], bins=50, edgecolor='white')
axes[0].set_title('SalePrice — raw (right-skewed)')
axes[0].set_xlabel('Sale Price ($)')

axes[1].hist(np.log1p(df['SalePrice']), bins=50, edgecolor='white', color='steelblue')
axes[1].set_title('log1p(SalePrice) — normalized')
axes[1].set_xlabel('log(Sale Price)')

plt.tight_layout()
plt.show()
print(f'Skewness (raw):    {df["SalePrice"].skew():.3f}')
print(f'Skewness (log1p):  {np.log1p(df["SalePrice"]).skew():.3f}')
```

```python
# Correlation heatmap — top 15 numerical features
num_df = df.select_dtypes(include=[np.number])
corr = num_df.corr()['SalePrice'].drop('SalePrice').abs().sort_values(ascending=False)
top15 = corr.head(15).index.tolist()

plt.figure(figsize=(12, 8))
sns.heatmap(
    num_df[top15 + ['SalePrice']].corr(),
    annot=True, fmt='.2f', cmap='Blues',
    linewidths=0.5, square=True
)
plt.title('Correlation matrix — top 15 features + SalePrice')
plt.tight_layout()
plt.savefig('../reports/figures/correlation_matrix_top15.png')
plt.show()
```

```python
# Missing value analysis
missing = df.isnull().sum()
missing = missing[missing > 0].sort_values(ascending=False)
print(f'Columns with missing values: {len(missing)}')

plt.figure(figsize=(12, 5))
missing.plot(kind='bar', color='coral', edgecolor='white')
plt.title('Missing value counts by column')
plt.xlabel('Column')
plt.ylabel('Missing count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# GrLivArea vs SalePrice — outlier identification
plt.figure(figsize=(10, 5))
plt.scatter(df['Gr Liv Area'], df['SalePrice'], alpha=0.4, edgecolors='none')

# Highlight the 2 documented outliers
outliers = df[(df['Gr Liv Area'] > 4000) & (df['SalePrice'] < 300000)]
plt.scatter(outliers['Gr Liv Area'], outliers['SalePrice'],
            color='red', s=80, label=f'Outliers to remove (n={len(outliers)})', zorder=5)

plt.xlabel('Gr Liv Area (sq ft)')
plt.ylabel('SalePrice ($)')
plt.title('Gr Liv Area vs SalePrice — outlier detection')
plt.legend()
plt.tight_layout()
plt.savefig('../reports/figures/grlivarea_outliers.png')
plt.show()

print('Outlier records:')
print(outliers[['Gr Liv Area', 'SalePrice', 'SaleCondition']].to_string())
```

```python
# Neighborhood median prices — bar chart
neigh_price = df.groupby('Neighborhood')['SalePrice'].median().sort_values(ascending=False)

plt.figure(figsize=(14, 5))
neigh_price.plot(kind='bar', color='steelblue', edgecolor='white')
plt.title('Median SalePrice by Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Median Sale Price ($)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
